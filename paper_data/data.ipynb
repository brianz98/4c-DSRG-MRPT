```python
def extract_spl_data_hidden(order):
    dcb_curve = []
    suffix = '' if order == 'pt3' else '-new'
    prefix = 'complete-curve' if order == 'pt3' else ''
    for file in glob.glob(f'{order}-curve/*{suffix}.out'):
        with open(file, 'r') as f:
            temp = f.readlines()
            dcb_curve.append([_.strip() for _ in temp])
    dcb_curve = [item for sublist in dcb_curve for item in sublist]

    data = {}
    key = None
    start = None
    idx_save = 0

    for idx, line in enumerate(dcb_curve):
        if 'NOW DOING' in line:
            idx_save = idx
            if (start != None):
                # not the first block
                data[key] = dcb_curve[start:idx]
            keystring = line.split('=')[-1].split('!')[0]
            key = round(float(keystring),4) if 'e' not in keystring else float(keystring)
            start = idx+1
    data[key] = dcb_curve[start:len(dcb_curve)]

    elements = ['B','C','N','O','F','Al','Si','P','S','Cl','Ga','Ge','As','Se','Br']
    length = [2,4,4,8,4,2,4,4,8,4,2,4,4,8,4]

    spl_curve = {}

    for key in data.keys():
        spl_curve[key] = {}
        iele = 0
        for idx,line in enumerate(data[key]):
            if f'Fully relaxed DSRG-MR{order.upper()} E_corr' in line:
                spl_curve[key][elements[iele]] = float(data[key][idx+2+length[iele]].split()[3])
                iele += 1
    
    return spl_curve

def extract_spl_data(order):
    dcb_curve = []
    suffix = '' if order == 'pt3' else '-new'
    prefix = 'complete-curve' if order == 'pt3' else ''
    for file in glob.glob(f'{order}-curve/*{suffix}.out'):
        with open(file, 'r') as f:
            temp = f.readlines()
            dcb_curve.append([_.strip() for _ in temp])
    dcb_curve = [item for sublist in dcb_curve for item in sublist]

    data = {}
    key = None
    start = None
    idx_save = 0

    for idx, line in enumerate(dcb_curve):
        if 'NOW DOING' in line:
            idx_save = idx
            if (start != None):
                # not the first block
                data[key] = dcb_curve[start:idx]
            keystring = line.split('=')[-1].split('!')[0]
            key = round(float(keystring),4) if 'e' not in keystring else float(keystring)
            start = idx+1
    data[key] = dcb_curve[start:len(dcb_curve)]

    elements = ['B','C','N','O','F','Al','Si','P','S','Cl','Ga','Ge','As','Se','Br']
    length = [2,1,10,5,4,2,1,8,5,4,2,1,8,5,4]

    spl_curve = {}

    for key in data.keys():
        spl_curve[key] = {}
        iele = 0
        for idx,line in enumerate(data[key]):
            if f'Fully relaxed DSRG-MR{order.upper()} E_corr' in line:
                spl_curve[key][elements[iele]] = float(data[key][idx+2+length[iele]].split()[3])
                iele += 1
    
    return spl_curve
```

```python
spl_curve_pt2 = pd.DataFrame(extract_spl_data('pt2'))
spl_curve_pt3 = pd.DataFrame(extract_spl_data('pt3'))
spl_curve_pt2_hidden = pd.DataFrame(extract_spl_data_hidden('pt2'))
spl_curve_pt2_hidden.rename(index={_:_+'_hidden' for _ in spl_curve_pt2_hidden.index}, inplace=True)
spl_curve_pt3_hidden = pd.DataFrame(extract_spl_data_hidden('pt3'))
spl_curve_pt3_hidden.rename(index={_:_+'_hidden' for _ in spl_curve_pt3_hidden.index}, inplace=True)

spl_curve_pt2_allpoints = pd.concat([spl_curve_pt2, spl_curve_pt2_hidden]).drop(index=['N_hidden','P_hidden','As_hidden','F_hidden','Cl_hidden','Br_hidden','B_hidden','Al_hidden','Ga_hidden'])
spl_curve_pt3_allpoints = pd.concat([spl_curve_pt3, spl_curve_pt3_hidden]).drop(index=['N_hidden','P_hidden','As_hidden','F_hidden','Cl_hidden','Br_hidden','B_hidden','Al_hidden','Ga_hidden'])

allpoints = pd.read_csv('allpoints.csv', index_col=0).drop(index=['N_hidden','P_hidden','As_hidden'])
spl_curve_pt2_allpoints[0.00] = allpoints[['CASSCF']]
spl_curve_pt3_allpoints[0.00] = allpoints[['CASSCF']]
```

```python
raw_df = pd.concat([allpoints,spl_curve_pt2_allpoints[0.2],spl_curve_pt2_allpoints[0.34],spl_curve_pt2_allpoints[0.5],spl_curve_pt3_allpoints[0.2],spl_curve_pt3_allpoints[0.34],spl_curve_pt3_allpoints[0.5]],axis=1)
raw_df.columns = ['Exp','CASSCF','CASPT2','MRCI','0.2','0.34','0.5','pt3_0.2','pt3_0.34','pt3_0.5']
signed_err = raw_df.copy()
signed_err = signed_err.drop('Exp',axis=1)
signed_err = signed_err - raw_df[['Exp']].values

signed_pct_err = raw_df.copy()
signed_pct_err - signed_pct_err.drop('Exp',axis=1)
signed_pct_err = 100*(signed_err/raw_df[['Exp']].values)
```
